```python
import torch
import numpy as np
from util.retrowrapper import RetroWrapper

game = 'IceClimber-Nes'
num_works = 4
envs = [RetroWrapper(game) for _ in range(num_works)]
```

```python
def get_action():
    action_dim = 9
    action_space = np.arange(action_dim)
    move = np.random.choice(action_space)
    binary_actions = torch.zeros(9, dtype=torch.int32)
    binary_actions[move] = 1.0
    action = binary_actions
    
    return action
```

```python
def work(env):
    for epo in range(20000):
        state = env.reset()
        for t in range(1000):
            action = get_action()
            print('action', action)
            env_res = env.step(action)
            if len(env_res) == 4:
                state_, reward, done, _ = env_res
            else:
                done = env_res
```

```python
import threading

for i in range(num_works):
    work_thread = threading.Thread(target=work, args=(envs[i], ))
    work_thread.start()
```
